##1. Cargar Libreria  

```python
import numpy as np
import pandas as pd
```

##2. Conexion a base de datos

```python
nxl="/content/sample_data/0. SolicitantesCrédito(Quiz 1).xlsx"
XDB= pd.read_excel(nxl, sheet_name=0)
```

##3. Llamar la variable de caracterizacion "edad"

```python
XDe=XDB[['Edad']]

XDe=XDe.dropna() #dropna remueve celdas vacias
XDe=np.array(XDe)
```

##4. Calcular desviacion estandar y media de la edad

```python
print("La media de la edad es:", np.mean(XDe))
print("La desviacion estandar de la variable es:", np.std(XDe))
print("la edad minima es:", np.mean(XDe)-np.std(XDe))
print("la edad maxima es:", np.mean(XDe)+np.std(XDe))
```

## 5. Histograma de frecuencias. Agrupar datos por intervalos

```python
import matplotlib.pyplot as plt

plt.hist(XDe, bins=15)
plt.show()

```

##6. Calcular valor de pertenencia para persona de 60 (1,0). Afinidad de un valor por una variable aleatoria

```python
VP= np.exp(-0.5*((49.12-np.mean(XDe))/np.std(XDe))**2)

print("El valor de pertenencia de una edad de 50 es:", VP)
```
